# Executable Code

```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt
import seaborn as sns
import missingno as msno
import datacompy as dc
sns.set()
from sklearn.model_selection import train_test_split
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
from sklearn.impute import KNNImputer
import io
import math
from sklearn.metrics import mean_squared_error
```

```python
pd.set_option("display.max_rows", None, "display.max_columns", None)
```

```python
incomplete_data = pd.read_excel('BCW_AN_1 - Incomplete Dataset.xlsx')
complete_data = pd.read_excel('BCW - Complete Dataset.xlsx')
```

```python
raw_data=incomplete_data
raw_data.describe()
raw_data.info()
```

```python
raw_data
```

```python
raw_data_with_header = pd.DataFrame(raw_data.values, columns = ["X"+str(i) for i in range(1,len(raw_data.columns)+1)])
raw_data_with_header
```

```python
# Check the missing value visualizations missingno doc  https://github.com/ResidentMario/missingno
msno.matrix(raw_data_with_header)
msno.heatmap(raw_data_with_header)
msno.dendrogram(raw_data_with_header)
```

```python
msno.bar(raw_data_with_header)
```

```python
original_data_xl=complete_data
```

```python
original_data_xl_with_header = pd.DataFrame(original_data_xl.values, columns = ["X"+str(i) for i in range(1,len(original_data_xl.columns)+1)])
original_data_xl_with_header
```

# ICKNNI

```python
imputer = KNNImputer(n_neighbors=1,weights="distance")
```

```python
# imputing in original dataset
z=imputer.fit_transform(raw_data_with_header)
icknn_imputed_data = pd.DataFrame(data=z)
```

```python
icknn_imputed_data 
```

```python
icknn_imputed_data_with_header = pd.DataFrame(icknn_imputed_data.values, columns = ["X"+str(i) for i in range(1,len(icknn_imputed_data.columns)+1)])
icknn_imputed_data_with_header 
```

```python
mse = mean_squared_error(original_data_xl_with_header, icknn_imputed_data_with_header)
rmse = math.sqrt(mse)
print(rmse)
```

# CCKNNI

```python
imputer = KNNImputer(n_neighbors=1,weights="distance")
```

```python
# doing calculation for only complete case
imputer.fit(raw_data.dropna())
```

```python
# imputing in original dataset
z=imputer.transform(raw_data)
```

```python
cknn_imputed_data = pd.DataFrame(data=z)
cknn_imputed_data_with_header=pd.DataFrame(cknn_imputed_data.values, columns = ["X"+str(i) for i in range(1,len(cknn_imputed_data.columns)+1)])
```

```python
cknn_imputed_data_with_header
```

```python
mse = mean_squared_error(original_data_xl_with_header, cknn_imputed_data_with_header)
```

```python
rmse = math.sqrt(mse)
print(rmse)
```

```python
icknn_imputed_data_with_header.to_excel('BCW_AN_1 - ICKNNI.xlsx')
cknn_imputed_data_with_header.to_excel('BCW_AN_1 - CCKNNI.xlsx')
```
